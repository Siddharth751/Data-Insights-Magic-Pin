```python
df['CREATED_DATE'] = pd.to_datetime(df['CREATED_DATE'])

# Calculate the current date (for recency calculation)
current_date = max(df['CREATED_DATE'])

# RFM analysis per user
rfm_df = df.groupby('USER_ID').agg({
    'CREATED_DATE': lambda x: (current_date - x.max()).days,  # Recency
    'TRANSACTION_ID': 'count',  # Frequency
    'AMOUNT': 'sum'  # Monetary
}).reset_index()

# Renaming columns for clarity
rfm_df.rename(columns={
    'CREATED_DATE': 'Recency',
    'TRANSACTION_ID': 'Frequency',
    'AMOUNT': 'Monetary'
}, inplace=True)

# Assign scores for each dimension (Recency, Frequency, Monetary)
rfm_df['Recency_Score'] = pd.qcut(rfm_df['Recency'], q=4, labels=False)
rfm_df['Frequency_Score'] = pd.qcut(rfm_df['Frequency'], q=4, labels=False)
rfm_df['Monetary_Score'] = pd.qcut(rfm_df['Monetary'], q=4, labels=False)

# Combine the scores into a single RFM Score
rfm_df['RFM_Score'] = rfm_df['Recency_Score'] * 100 + rfm_df['Frequency_Score'] * 10 + rfm_df['Monetary_Score']

# Create RFM segments based on scores
def rfm_segment(row):
    if row['RFM_Score'] >= 300:
        return 'High Value'
    elif 200 <= row['RFM_Score'] < 300:
        return 'Mid Value'
    else:
        return 'Low Value'

rfm_df['RFM_Segment'] = rfm_df.apply(rfm_segment, axis=1)

# Print the RFM DataFrame with segments
print(rfm_df[['USER_ID', 'Recency', 'Frequency', 'Monetary', 'RFM_Score', 'RFM_Segment']])
```

```python
import pandas as pd
import datetime as dt
import matplotlib.pyplot as plt

# Assuming you have a DataFrame 'df' with the relevant fields

# Convert 'CREATED_DATE' to datetime
df['CREATED_DATE'] = pd.to_datetime(df['CREATED_DATE'])

# Calculate the current date (for recency calculation)
current_date = max(df['CREATED_DATE'])

# RFM analysis per user
rfm_df = df.groupby('USER_ID').agg({
    'CREATED_DATE': lambda x: (current_date - x.max()).days,  # Recency
    'TRANSACTION_ID': 'count',  # Frequency
    'AMOUNT': 'sum'  # Monetary
}).reset_index()

# Renaming columns for clarity
rfm_df.rename(columns={
    'CREATED_DATE': 'Recency',
    'TRANSACTION_ID': 'Frequency',
    'AMOUNT': 'Monetary'
}, inplace=True)

# Assign scores for each dimension (Recency, Frequency, Monetary)
rfm_df['Recency_Score'] = pd.qcut(rfm_df['Recency'], q=4, labels=False)
rfm_df['Frequency_Score'] = pd.qcut(rfm_df['Frequency'], q=4, labels=False)
rfm_df['Monetary_Score'] = pd.qcut(rfm_df['Monetary'], q=4, labels=False)

# Combine the scores into a single RFM Score
rfm_df['RFM_Score'] = rfm_df['Recency_Score'] * 100 + rfm_df['Frequency_Score'] * 10 + rfm_df['Monetary_Score']

# Create RFM segments based on scores
def rfm_segment(row):
    if row['RFM_Score'] >= 300:
        return 'High Value'
    elif 200 <= row['RFM_Score'] < 300:
        return 'Mid Value'
    else:
        return 'Low Value'

rfm_df['RFM_Segment'] = rfm_df.apply(rfm_segment, axis=1)

# Create a bar chart for RFM segments
segment_counts = rfm_df['RFM_Segment'].value_counts().sort_index()

plt.figure(figsize=(10, 6))
segment_counts.plot(kind='bar', color='skyblue')
plt.xlabel("RFM Segment")
plt.ylabel("Number of Users")
plt.title("Number of Users in Each RFM Segment")
plt.xticks(rotation=45)
plt.grid(axis='y')

# Display the count values on top of each bar
for i, count in enumerate(segment_counts):
    plt.text(i, count, str(count), ha='center', va='bottom')

plt.tight_layout()
plt.show()
```

```python
def calculate_and_convert_currency(df):
    # Calculate the sum of amounts in GBP and Euro
    gbp_sum = df[df['CURRENCY'] == 'GBP']['AMOUNT'].sum()
    euro_sum = df[df['CURRENCY'] == 'EUR']['AMOUNT'].sum()

    # Define the exchange rate (for example, 1 GBP = 1.15 EUR)
    exchange_rate = 1.15

    # Convert GBP to Euro for a direct comparison
    gbp_to_euro = gbp_sum * exchange_rate

    return gbp_sum, euro_sum, gbp_to_euro

# Call the function and print the results
gbp_sum, euro_sum, gbp_to_euro = calculate_and_convert_currency(df)

print("Sum of GBP Amounts:", gbp_sum)
print("Sum of Euro Amounts:", euro_sum)
print("GBP to Euro Conversion:", gbp_to_euro)
```

```python
def calculate_currency_transactions(df):
    # Calculate the number of transactions in GBP and Euro
    gbp_transactions = df[df['CURRENCY'] == 'GBP']['CURRENCY'].count()
    euro_transactions = df[df['CURRENCY'] == 'EUR']['CURRENCY'].count()
    
    return gbp_transactions, euro_transactions

# Call the function and print the results
gbp_transactions, euro_transactions = calculate_currency_transactions(df)

print("Number of GBP Transactions:", gbp_transactions)
print("Number of Euro Transactions:", euro_transactions)
```
